```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criar dados fictícios
data = {
    'Dia': pd.date_range(start='2024-01-01', periods=7),
    'Horas_Trabalhadas': [8, 7, 8, 9, 6, 8, 7],
    'Bugs_Corrigidos': [3, 2, 4, 1, 0, 2, 1],
    'Tarefas_Concluidas': [10, 8, 12, 9, 11, 7, 10],
}

df = pd.DataFrame(data)
df['Produtividade'] = df['Tarefas_Concluidas'] / df['Horas_Trabalhadas']
df['Media_Horas_Trabalhadas'] = df['Horas_Trabalhadas'].mean()
df['Media_Bugs_Corrigidos'] = df['Bugs_Corrigidos'].mean()
df['Media_Tarefas_Concluidas'] = df['Tarefas_Concluidas'].mean()
df['Media_Produtividade'] = df['Produtividade'].mean()

df
```

```python
# Gráfico de Horas Trabalhadas
plt.figure(figsize=(10, 5))
plt.plot(df['Dia'], df['Horas_Trabalhadas'], marker='o', linestyle='-', color='b')
plt.title('Horas Trabalhadas ao longo da Semana')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.grid(True)
plt.show()

# Gráfico de Bugs Corrigidos
plt.figure(figsize=(10, 5))
plt.plot(df['Dia'], df['Bugs_Corrigidos'], marker='o', linestyle='-', color='g')
plt.title('Bugs Corrigidos ao longo da Semana')
plt.xlabel('Dia')
plt.ylabel('Bugs Corrigidos')
plt.grid(True)
plt.show()

# Gráfico de Tarefas Concluídas
plt.figure(figsize=(10, 5))
plt.plot(df['Dia'], df['Tarefas_Concluidas'], marker='o', linestyle='-', color='r')
plt.title('Tarefas Concluídas ao longo da Semana')
plt.xlabel('Dia')
plt.ylabel('Tarefas Concluídas')
plt.grid(True)
plt.show()

# Gráfico de Produtividade
plt.figure(figsize=(10, 5))
plt.plot(df['Dia'], df['Produtividade'], marker='o', linestyle='-', color='purple')
plt.title('Produtividade ao longo da Semana')
plt.xlabel('Dia')
plt.ylabel('Produtividade (Tarefas/Hora)')
plt.grid(True)
plt.show()
```
